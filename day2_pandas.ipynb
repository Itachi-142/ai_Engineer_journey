```python
# Pandas can read CSVs directly from a URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# First look — always do these four things when you see new data
print(df.shape)           # how many rows, columns
print(df.head())          # first 5 rows
print(df.info())          # column names, types, non-null counts
print(df.describe())      # statistics for numerical columns
```

```python
# Select a single column — returns a Series
print(df['Age'])

# Select multiple columns — returns a DataFrame
print(df[['Name', 'Age', 'Survived']])

# Select rows by condition — this is the most important one
survivors = df[df['Survived'] == 1]
print(f"Survivors: {len(survivors)}")

# Combine conditions
young_survivors = df[(df['Survived'] == 1) & (df['Age'] < 18)]
print(f"Young survivors (under 18): {len(young_survivors)}")

# .loc — select by label/condition
# .iloc — select by integer position
print(df.iloc[0])          # first row
print(df.loc[0:4, ['Name', 'Age', 'Survived']])  # rows 0-4, specific columns
```

```python
# Check missing values
print(df.isnull().sum())

# Which percentage of each column is missing?
print((df.isnull().sum() / len(df)) * 100)

# Three strategies:
# 1. Drop rows with missing values
df_dropped = df.dropna()
print("After dropping all nulls:", df_dropped.shape)

# 2. Fill with mean (for numerical columns)
df['Age'] = df['Age'].fillna(df['Age'].mean())

# 3. Fill with mode (for categorical columns)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Verify
print(df.isnull().sum())
```

```python
# Survival rate by gender
print(df.groupby('Sex')['Survived'].mean())

# Average age by passenger class
print(df.groupby('Pclass')['Age'].mean())

# Multiple aggregations at once
print(df.groupby('Pclass').agg({
    'Survived': 'mean',
    'Age': 'mean',
    'Fare': 'mean'
}).round(2))
```

```python
# Create a new column
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# Create a categorical column from a condition
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)

# Apply a custom function to a column
def age_group(age):
    if age < 18:
        return 'child'
    elif age < 60:
        return 'adult'
    else:
        return 'senior'

df['AgeGroup'] = df['Age'].apply(age_group)
print(df[['Age', 'AgeGroup', 'FamilySize', 'IsAlone']].head(10))
```

```python
survival_rate = df.groupby("Pclass")["Survived"].mean()

print("Survival rate by class:")
print(survival_rate)

print("\nClass with highest survival:",
      survival_rate.idxmax())

avg_fare = df.groupby("Survived")["Fare"].mean()

print("\nAverage fare (0=dead, 1=survived):")
print(avg_fare)

df["FarePerPerson"] = df["Fare"] / df["FamilySize"]


print("\nEmbarked counts:")
print(df["Embarked"].value_counts())

filtered = df[
    (df["Sex"] == "female") &
    (df["Age"] > 30) &
    (df["Survived"] == 1)
]

print("\nFemale > 30 & survived count:", len(filtered))

df = df.drop(columns=["Cabin", "Ticket", "Name"])

print("\nRemaining columns:")
print(df.columns)
```
